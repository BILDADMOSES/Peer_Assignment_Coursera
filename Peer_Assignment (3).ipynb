```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
df = pd.read_csv("cardata.csv") #Reading the csv file
df.dropna(inplace = True) #Removing rows with empty values
nw_df = df[df["Selling_Price"] > 3.00 ]
nw_df
sp = nw_df["Selling_Price"] 
km = nw_df["Kms_Driven"]
fig = plt.figure(figsize = (12,10))
plt.title("Relationship Between Km Driven and Selling Price")
plt.xlabel("Kms_Driven")
plt.ylabel("Selling_price")

km_sp = np.column_stack((km,sp))
km_sp
k_res = KMeans(n_clusters= 3).fit(km_sp)
cluster = k_res.cluster_centers_
cluster
plt.scatter(km,sp)
plt.scatter(cluster[:,0],cluster[:,1],s=1000,alpha=.70)

plt.grid()
plt.show()
```
